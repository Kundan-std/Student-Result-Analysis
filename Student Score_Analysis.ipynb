```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv("Students_Scores.csv")
print(df. head())
```

#Drop unnamed column

```python
df = df.drop("Unnamed: 0",axis = 1)
print(df.head())
```

```python
df.head()
```

#Gender Distribution

```python
sns.countplot(data = df,x="Gender")
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.countplot(data = df,x="Gender")
plt.show()
```

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(data = df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

#from the above chart we have analysed that:
#the number of females in the data is more then the males

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb , annot=True)
plt.title("Relationship between parint's Education and student's scores ")
plt.show()
```

#from the above chart we have concluded that education of the parents have a good impact on there scores
